```python
import requests
from bs4 import BeautifulSoup

# 논문 구분(학위, 학술지), 논문명, 연도, 초록 요약, detail_links 
keyword='자연어'
inner_links=[]
deg_links=[]
thesis_infos=[]
        
for deg_page in range(33):
    deg_the='''http://www.riss.kr/search/Search.do?detailSearch=false&searchGubun=true&query='''+keyword+'''
              &colName=bib_t&pageScale=20&iStartCount='''+str(20*deg_page)+'''&searchGubun=true'''

    req_deg=requests.get(deg_the).content
    soup_deg=BeautifulSoup(req_deg, 'html.parser')

    deg_lists=soup_deg.select('.degree li .txt')
    deg_etc=soup_deg.select('.degree li .etc')

    for title, year in zip(deg_lists, deg_etc):
        dic2={
            'subpage_link':'http://www.riss.kr'+title.select('a')[0].get('href'),
            'thesis_title':title.text.replace('\n', ''),
            'thesis_year':year.text.split('[')[1].split(']')[0],
            'Gubun':'학위논문'
        }
        deg_link='http://www.riss.kr'+title.select('a')[0].get('href')
        deg_title=title.text.replace('\n', '')
        deg_year=year.text.split('[')[1]
        deg_year=deg_year.split(']')[0]       
        thesis_infos.append(dic2)
        deg_links.append(deg_link)
# print(len(inner_links))
print(len(deg_links))
```

```python
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import time

# path = 'c:\myPycode/mini_project/chromedriver.exe'

wb = webdriver.Chrome(ChromeDriverManager().install())
inner_absb=[]

for sub_link in deg_links:
    wb.get(sub_link)
    time.sleep(1)
    try:
        a=[]
        a.append(wb.find_elements_by_css_selector('#btnabs1'))
        for i in a:
            i[0].click()
    except Exception as e:
        print(e)
        print('no1')
    try:
        b=[]
        b.append(wb.find_elements_by_css_selector('#btnabs2'))
        for j in b:
            j[0].click()
    except Exception as e2:
        print(e2)
        print('no2')
    time.sleep(1)
    texts=wb.find_elements_by_css_selector('.tt')
    sub_abs=''
    for text in texts: 
        sub_abs=sub_abs+str(text.text)
    inner_absb.append(sub_abs)
```
